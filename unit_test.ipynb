```python
# í…ŒìŠ¤íŠ¸ ê²°ê³¼ ì €ì¥ìš©
test_results = []

def run_test(test_name, test_func, *args, **kwargs):
    """í…ŒìŠ¤íŠ¸ í•¨ìˆ˜ ì‹¤í–‰ ë° ê²°ê³¼ ê¸°ë¡"""
    print(f"\nğŸ§ª í…ŒìŠ¤íŠ¸: {test_name}")
    print("-" * 50)
    
    try:
        result = test_func(*args, **kwargs)
        print(f"âœ… {test_name} - ì„±ê³µ")
        test_results.append({"test": test_name, "status": "PASS", "error": None})
        return result
    except Exception as e:
        print(f"âŒ {test_name} - ì‹¤íŒ¨: {str(e)}")
        print(f"ğŸ” ìƒì„¸ ì˜¤ë¥˜:\n{traceback.format_exc()}")
        test_results.append({"test": test_name, "status": "FAIL", "error": str(e)})
        return None

def expect_error(test_name, test_func, expected_error, *args, **kwargs):
    """ì—ëŸ¬ ë°œìƒì„ ê¸°ëŒ€í•˜ëŠ” í…ŒìŠ¤íŠ¸"""
    print(f"\nğŸ§ª ì—ëŸ¬ í…ŒìŠ¤íŠ¸: {test_name}")
    print("-" * 50)
    
    try:
        result = test_func(*args, **kwargs)
        print(f"âŒ {test_name} - ì‹¤íŒ¨: ì˜ˆìƒëœ ì—ëŸ¬ê°€ ë°œìƒí•˜ì§€ ì•ŠìŒ")
        test_results.append({"test": test_name, "status": "FAIL", "error": "Expected error not raised"})
        return False
    except expected_error as e:
        print(f"âœ… {test_name} - ì„±ê³µ: ì˜ˆìƒëœ ì—ëŸ¬ ë°œìƒ ({type(e).__name__}: {str(e)})")
        test_results.append({"test": test_name, "status": "PASS", "error": None})
        return True
    except Exception as e:
        print(f"âŒ {test_name} - ì‹¤íŒ¨: ì˜ˆìƒê³¼ ë‹¤ë¥¸ ì—ëŸ¬ ë°œìƒ ({type(e).__name__}: {str(e)})")
        test_results.append({"test": test_name, "status": "FAIL", "error": f"Unexpected error: {str(e)}"})
        return False

def create_test_csv(filename, data):
    """í…ŒìŠ¤íŠ¸ìš© CSV íŒŒì¼ ìƒì„±"""
    df = pd.DataFrame(data)
    df.to_csv(filename, index=False, encoding='utf-8')
    return filename

print("ğŸ”§ í…ŒìŠ¤íŠ¸ í—¬í¼ í•¨ìˆ˜ ì„¤ì • ì™„ë£Œ")
```
